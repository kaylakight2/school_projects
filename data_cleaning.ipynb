```python
# Open the metadata file in read mode
with open("census_income_metadata.txt", "r") as file:
    # Read the contents of the file
    text = file.read()

lines = text.split("\n")

lines

#column_names = []

#for line in lines:
  #  if "|" in line:
  #      parts = line.split("|")
  #      column_name = parts[0].strip()
  #      abbreviated_name = parts[1].strip()
  #      column_names.append((column_name, abbreviated_name))
        
#column_names

column_names = []
for line in lines:
    if '\t' in line:
        column_name = line[:line.index('\t')].strip()
        abbreviated_name = line[line.index('\t')+1:].strip()
        column_names.append((column_name, abbreviated_name))


column_names
```

```python
column_names = column_names[:-5]
column_names
```

```python
column_dict = {}
for column in column_names:
    key = column[0].strip('| ').title()  # extract and format the first part of the line
    value = column[1]  # extract the second part of the line
    column_dict[key] = value

print(column_dict)
```

```python
abbrev_list = list(column_dict.values())
print(abbrev_list)
```

```python
metadf = pd.DataFrame.from_dict(column_dict, orient='index', columns=['var_code'])

metadf.reset_index(inplace=True)
metadf.index += 1
metadf.columns = ['variable', 'var_code']
metadf
```

```python
lines

pattern = r'#\d+ \((.*?)\) (\w+)'
attributes = {}

for line in lines:
    match = re.search(pattern, line)
    if match:
        attribute_name = match.group(1)
        attribute_type = match.group(2)
        attributes[attribute_name] = attribute_type
        
print(attributes)
```

```python
df_data = pd.DataFrame.from_dict(attributes, orient='index', columns=['data_type'])
df_data.reset_index(inplace=True)
df_data.index += 1
df_data.columns = ['variable', 'data_type']


df_data
```

```python
def set_data_type(df):
    for index, row in df.iterrows():
        data_type = row['data_type']
        if pd.isna(data_type):
            df.at[index, 'variable'] = df.at[index, 'variable'].astype(str)
        elif data_type == 'nominal':
            df.at[index, 'variable'] = df.at[index, 'variable'].astype('category')
        elif data_type == 'continuous':
            df.at[index, 'variable'] = df.at[index, 'variable'].astype('float64')
    return df
```
